```python
%pylab inline
import scipy.io.netcdf
```

```python
# Get re-gridded data, gridded to (88,80) grid which
# is the medium size grid
high_res = scipy.io.netcdf_file('high/prog__0001_006.nc') # not re-gridded
med_res = scipy.io.netcdf_file('med/prog__0001_006.nc') # not re-gridded
low_res = scipy.io.netcdf_file('low/prog__0001_006.nc') # not re-gridded

# remap to medium res grid
high_res_med_grid = scipy.io.netcdf_file('high/prog_high_med_remap.nc')
low_res_med_grid = scipy.io.netcdf_file('low/prog_low_med_remap.nc')
```

```python
high_e = high_res_med_grid.variables['e']
med_e = med_res.variables['e']
low_e = low_res_med_grid.variables['e']
```

```python
# remap from low and high resolution to medium resolution grid

fig, (ax0, ax1, ax2) = plt.subplots(nrows=3)

# adjust figures for viewing purposes
fig.subplots_adjust(hspace=.3, bottom=8, top=10)


# High resolution
high_cf = ax0.contourf(high_e[0,0], cmap=cm.nipy_spectral)
fig.colorbar(high_cf, ax=ax0)
ax0.set_title('(196, 180) -> (88, 80)')
ax0.margins(x=.5, y=.5)


# medium resolution
med_cf = ax1.contourf(med_e[0,0], cmap=cm.nipy_spectral)
ax1.set_title('(88, 80)')
fig.colorbar(med_cf, ax=ax1)

# Low resolution
low_cf = ax2.contourf(low_e[0,0], cmap=cm.nipy_spectral)
ax2.set_title('(44, 40) -> (88, 80)')
fig.colorbar(low_cf, ax=ax2)


plt.show()
```

```python
# Get re-gridded data, gridded to (44, 40) grid which
# is the low res size grid
high_res_low_grid = scipy.io.netcdf_file('high/prog_high_low_remap.nc')
med_res_low_grid = scipy.io.netcdf_file('med/prog_med_low_remap.nc')
```

```python
high_e = high_res_low_grid.variables['e']
med_e = med_res_low_grid.variables['e']
low_e = low_res.variables['e']
```

```python
# remap from high and med to low resolution grid

fig, (ax0, ax1, ax2) = plt.subplots(nrows=3)

# adjust figures for viewing purposes
fig.subplots_adjust(hspace=.3, bottom=8, top=10)


# High resolution
high_cf = ax0.contourf(high_e[0,0], cmap=cm.nipy_spectral)
fig.colorbar(high_cf, ax=ax0)
ax0.set_title('(196, 180) -> (44, 40)')
ax0.margins(x=.5, y=.5)


# medium resolution
med_cf = ax1.contourf(med_e[0,0], cmap=cm.nipy_spectral)
ax1.set_title('(88, 80) -> (44, 40)')
fig.colorbar(med_cf, ax=ax1)

# Low resolution
low_cf = ax2.contourf(low_e[0,0], cmap=cm.nipy_spectral)
ax2.set_title('(44, 40)')
fig.colorbar(low_cf, ax=ax2)


plt.show()
```

```python
# same remap to low resolution grid just using pcolormesh instead of contourf

fig, (ax0, ax1, ax2) = plt.subplots(nrows=3)

# adjust figures for viewing purposes
fig.subplots_adjust(hspace=.3, bottom=8, top=10)


# High resolution
high_cf = ax0.pcolormesh(high_e[0,0], cmap=cm.nipy_spectral)
ax0.set_title('(196, 180) -> (44, 40)')
fig.colorbar(high_cf, ax=ax0)
ax0.margins(x=.5, y=.5)


# medium resolution
med_cf = ax1.pcolormesh(med_e[0,0], cmap=cm.nipy_spectral)
ax1.set_title('(88, 80) -> (44, 40)')
fig.colorbar(med_cf, ax=ax1)

# Low resolution
low_cf = ax2.pcolormesh(low_e[0,0], cmap=cm.nipy_spectral)
ax2.set_title('(44, 40)')
fig.colorbar(low_cf, ax=ax2)


plt.show()
```
