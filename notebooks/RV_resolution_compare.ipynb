```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size':16})
```

```python
datapath = '/raid/ra40/data/ras/double_gyre/'
resolutions = ['2.5km','5km','10km','20km','40km']
data = {}
for res in resolutions:
    data[res] = xr.open_dataset(datapath+res+'/'+'prog__0001_006.nc',decode_times=False)
```

```python
res = '2.5km'
plt.figure(figsize=(28,12))
counter = 1
for res in resolutions:
    plt.subplot(2,3,counter)
    plt.pcolormesh(data[res].xq,data[res].yq,data[res].RV[-1,0,:,:]*1e6,cmap=plt.cm.RdBu_r,vmin=-10,vmax=10)
    cax = plt.colorbar()
    cax.set_label('Relative Vorticity [$10^{-6}/s$]')
    plt.xlim(2,11); plt.ylim(37.5,42.5)
    counter += 1
    plt.title(res)
```
